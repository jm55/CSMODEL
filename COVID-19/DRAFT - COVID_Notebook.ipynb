## Earliest and Latest Dates

```python
date_values = covid_df['date'].unique()
date_values = np.sort(date_values,kind='mergesort')
dateCount = date_values.size
print("dateCount:",dateCount,"\n")
print("earliest:",date_values[0])
print("latest (valid):",date_values[len(date_values)-1])
#the latest possible data maybe incomplete, so to be safe let's just adjust it to the day before the latest date entered
```

## Separating Countries to Individual Dataframes

```python
#countries to individual dataframes
world_df = covid_df[covid_df['iso_code']=='OWID_WRL'] #World
phl_df = covid_df[covid_df['iso_code']=='PHL'] #Philippines
brn_df = covid_df[covid_df['iso_code']=='BRN'] #Brunei
khm_df = covid_df[covid_df['iso_code']=='KHM'] #Cambodia
idn_df = covid_df[covid_df['iso_code']=='IDN'] #Indonesia
sgp_df = covid_df[covid_df['iso_code']=='SGP'] #Singapore
lao_df = covid_df[covid_df['iso_code']=='LAO'] #Lao PDR
tha_df = covid_df[covid_df['iso_code']=='THA'] #Thailand
mys_df = covid_df[covid_df['iso_code']=='MYS'] #Malaysia
mmr_df = covid_df[covid_df['iso_code']=='MMR'] #Myanmar
vnm_df = covid_df[covid_df['iso_code']=='VNM'] #Vietnam

#compiled (but separated) dataframes
compiled_str = ['phl_df','brn_df','khm_df','idn_df','sgp_df','lao_df','tha_df','mys_df','mmr_df','vnm_df','world_df']
compiled = [phl_df,brn_df,khm_df,idn_df,sgp_df,lao_df,tha_df,mys_df,mmr_df,vnm_df,world_df] #lists of lists, hehehe

#number of observations and sorting by 'date'
print("entries/observations/daysWithCOVID per country:")
for i in range(len(compiled)):
    print(compiled_str[i],':',compiled[i].shape[0])
    compiled[i].sort_values(by='date',kind='mergesort')
print("")

print("earliest records by country: ")
for i in range(len(compiled)):
    print(compiled_str[i],":",compiled[i]['date'].iloc[0])
print("\n")

print("latest records by country: ")
for i in range(len(compiled)):
    print(compiled_str[i],":",compiled[i]['date'].iloc[compiled[i].shape[0]]-1)
```
