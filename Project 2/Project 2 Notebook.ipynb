### 1. Dataset Representation

The dataset that was chosen for this project was *Dataset3.csv*. It could be desribed that the dataset shows what seems like **an association type of problem**, where **each column may represent an item or word from a store or a document, with each row possibly consisting either a customer or document.**

```python
import numpy as np #NUMPY
import pandas as pd #PANDAS
import matplotlib.pyplot as plt #MATPLOTLIB
```

```python
main_df = pd.read_csv('Dataset3.csv')
main_df
```

```python
main_df.info()
```

```python
#Removing 'Unnamed 0:'
main_df.drop(columns='Unnamed: 0', inplace=True)
main_df
```

### 2. Exploratory Data Analysis

```python
vis_df = main_df.iloc[:, 1:main_df.shape[1]] #FOR USE FOR NOTEBOOK VISUALIZATION
```

```python
totals = vis_df.sum() #totals by column
means = vis_df.mean() #means by column
```

```python
print('totals mean all-throughout: {0:.0f}'.format(totals.mean()))
print('mean all-throughout: {0:.0f}'.format(means.mean()))
```

It suggests that, on average, all words/items have a mean total of 250 and an average count of 3 regardless of document/customer.

```python
totals.plot.barh(x='column',y='total',figsize=[15,100],title='Total Count by Column (Item/Word)')
```

```python
means.plot.barh(x='column',y='total',figsize=[15,100],title='Average Count by Column (Item/Word)')
```

The bar graph shows that there that the counts of each item/word is relatively close to each other.
