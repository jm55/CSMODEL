### 2. Exploratory Data Analysis

```python
vis_df = main_df.iloc[:, 1:main_df.shape[1]] #FOR USE FOR NOTEBOOK VISUALIZATION
```

```python
totals = vis_df.sum() #totals by column
means = vis_df.mean() #means by column
```

```python
print('totals median:\n{0:.0f}'.format(totals.sort_values().median()))
print('means median:\n{0:.0f}'.format(means.sort_values().median()))
```

```python
totals.plot.barh(x='column',y='total',figsize=[15,100],title='Totals by Column (Item/Word)')
```

```python
means.plot.barh(x='column',y='total',figsize=[15,100],title='Totals by Column (Item/Word)')
```

The bar graph shows that there that the counts of each item/word is relatively close to each other.
